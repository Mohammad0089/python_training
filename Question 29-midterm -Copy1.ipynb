```python
import pandas as pd
import numpy as np
df = np.array([['Toyota Corolla', '40', '20175'],['Ford', '45', '25000'],['Dodge', '62', '35782'],['Chevrolet', '50', '30000'],['Canoo', '57', '34750'],['Tesla', '113', '54000'],['BMW', '70', '36400']])
df = pd.DataFrame(df)
df.head(10)
```

```python
df.rename(columns={0:'Car', 1:'Mileage', 2:'Cost'}, inplace=True)
df.columns
y = df.iloc[:, 2]
x = df.iloc[:, :2]
x.head()
```

```python
x = pd.get_dummies(x, columns=['Car'])
x.head(10)
```

```python
x['Mileage'] = x['Mileage'].astype(int)
y = y.astype(int)


from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor()
model.fit(x,y)

from sklearn import tree
from matplotlib import pyplot as plt
fig = plt.figure(figsize=(15,10))
_ = tree.plot_tree(model, feature_names=x.columns, filled=True)
```
