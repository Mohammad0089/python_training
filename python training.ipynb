```python
class Stack():
    def __init__(self):
        self.s = []
   
    def push(self, i ):
        self.s.append(i)
        
    def pop(self):
        return self.s.pop()
    
    def peek(self):
        return self.s[-1]
    
    def is_empty(self):
        if self.s == []:
            return True
        else:
            return False
        
    def getStack(self):
        return self.s
    
    def getLength(self):
        return len(self.s)

number = 70
s = Stack()
def convert_int_to_bin(dec_num):
    while dec_num > 0:
        remainder = dec_num % 2
        s.push(remainder)
        dec_num = dec_num // 2

    bin_num = ""
    while not s.is_empty():
        bin_num += str(s.pop())

    return bin_num
```
